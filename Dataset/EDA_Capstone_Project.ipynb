# EDA

```python
#Import all of the library that we're going to use
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns 
```

## Instagram

```python
#Read Instagram dataset
Instagram = pd.read_csv("https://raw.githubusercontent.com/fachrul-rh/influeHands-ML/main/Dataset/Instagram.csv")
```

```python
#Returns the first 5 rows for Instragram Dataset
Instagram.head()
```

```python
#See the size of our Instagram Dataset
Instagram.shape
```

```python
#See the info and dtype from Instagram Dataset
Instagram.info()
```

```python
#See if there's a NaN element in Instagram Dataset
Instagram.isna().sum()
#There is 7 missing items on Column 'Nama'
#But it will not be much problem for the future
#So just let it be
```

```python
#See statistically the info obtained from Instagram Dataset
Instagram.describe()
```

```python
#See the correlation between features in Instagram Dataset
plt.figure(figsize=(13,8))

correlation_matrix = Instagram.corr().round(2)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(data = correlation_matrix, mask=mask, annot = True)

plt.show()
```

```python
#See the skewness of each feature in the Instagram Dataset
Instagram.hist(figsize=(20,15))
#The feature "Fashion" and "Entertainment" has the amount of data exceeds 20
```
